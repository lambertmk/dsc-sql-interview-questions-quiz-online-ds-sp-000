## Question 2

Explain the relationship between **Primary Keys** and **Foreign Keys**.

Type your answer below this line:
_______________________________________________________________________________________________________________________________
A primary key is a unique identifier in a table. This would not be shared by rows within that table. To use a real life example, a primary key would be a driver's license number. This doesn't always have to be "usable" - it could be an sequential integer used as an index.
A foreign key is a piece of information shared between tables. These allow us to JOIN between tables. The "foreign" comes from that use; it allows us to "connect" the data between tables, and pull in based on that key.

## Question 4

Explain the various types of JOINs possible with SQL. 

Type your answer below this line:
_______________________________________________________________________________________________________________________________
LEFT JOIN pulls all information from the left table and overlapping info from right
JOIN/INNER JOIN pulls information that overlaps (key used for join is not null)
RIGHT JOIN pulls all information from the right table, overlapping info from left
FULL OUTER JOIN pulls all info from both tables, even where nulls are present
